## adding styles and colors

```python
def pdf(X, mu, sigma):
    a = 1. /(sigma*np.sqrt(2.*np.pi))
    b = -1. /(2.*sigma**2)
    return a*np.exp(b*(X-mu)**2)

X = np.linspace(-6, 6, 1000)

for i in range(5):
    samples = np.random.standard_normal(50)
    mu, sigma = np.mean(samples), np.std(samples)
    plt.plot(X, pdf(X, mu, sigma), color = '.75')
    
plt.plot(X, pdf(X, 0., 1.), color = 'k')
plt.show()
```

## Lab: Scatter Plot Grey Scale From a File

```python
label_set = (  b'Iris-setosa',  b'Iris-versicolor',  b'Iris-virginica')
```

```python
def read_label(label):  
    return label_set.index(label)
```

```python
label_set
```

```python
data = np.loadtxt('iris.txt',                             
                  delimiter = ',',                             
                  converters = { 4 : read_label })
```

```python
color_set = ('.20', '.55', '.90') 
color_list = [color_set[int(label)] 
              for label in data[:,4]]

plt.scatter(data[:,0], data[:,1], color = color_list) 
plt.show()
```

```python
color_set = ('r', 'b', 'm') 
color_list = [color_set[int(label)] 
              for label in data[:,4]]

plt.scatter(data[:,0], data[:,1], color = color_list) 
plt.show()
```

## Lab: Markers - Controlling Marker Size

```python
A = np.random.standard_normal((100, 2)) 
A += np.array((-1, -1))

B = np.random.standard_normal((100, 2)) 
B += np.array((1, 1))

plt.scatter(B[:,0], B[:,1], c = '.65', s = 100.) 
plt.scatter(A[:,0], A[:,1], c = '.30', s = 25.)
plt.show() 
```

## Lab: Controlling Marker Frequency

```python
X = np.linspace(-6, 6, 1024) 
Y1 = np.sinc(X) 
Y2 = np.sinc(X) + 1

plt.plot(X, Y1, marker = 'o', color = '.45',markevery = 25) 
plt.plot(X, Y2, marker = 'o', color = 'k', markevery = 50)
plt.show()
```

## Lab: List as Input for Size Parameter

```python
M = np.random.standard_normal((1000, 2)) 
R = np.sum(M ** 2, axis = 1)

plt.scatter(M[:, 0], M[:, 1], c = 'b', marker = 's', s = 10 * R) 
plt.show()
```

## Labeling our Ticks

```python
import matplotlib.ticker as ticker 
```

```python
name_list = ('Mike', 'Brad', 'Stan', 'Jon', 'Matt') 
value_list = np.random.randint(0, 99, size = len(name_list)) 
pos_list = np.arange(len(name_list))

ax = plt.axes() 
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list))) 
ax.xaxis.set_major_formatter(ticker.FixedFormatter((name_list)))

plt.bar(pos_list, value_list, color = '.75', align = 'center') 
plt.show()
```

## Labeling our Ticks -the easy way

```python
name_list = ('Mike', 'Brad', 'Stan', 'Jon', 'Matt') 
value_list = np.random.randint(0, 99, size = len(name_list)) 
pos_list = np.arange(len(name_list))

plt.bar(pos_list, value_list, color = '.75', align = 'center')
plt.xticks(pos_list, name_list)
plt.show()
```

## Exploring the Sundry Color Schemes

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data)
plt.title("whitegrid")
plt.show()

sns.set_style("darkgrid")
sns.boxplot(data=data);
plt.title("darkgrid")
plt.show()

sns.set_style("white")
sns.boxplot(data=data);
plt.title("white")
plt.show()

sns.set_style("dark")
sns.boxplot(data=data);
plt.title("dark")
plt.show()

sns.set_style("ticks")
sns.boxplot(data=data);
plt.title("ticks")
plt.show()
# the 2 here are equal
sns.boxplot(data=data);
plt.title("ticks")
plt.show()
```
